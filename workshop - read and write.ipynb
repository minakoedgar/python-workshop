```python
# Read Text
def read_text (intext):
    infile = open(intext, 'rb')
    d = {}
    i = 1
    for line in infile:
        d[i] = line.split(',')
        i = i + 1
    return d

d_txt = read_text(intext)
print len(d_txt)
print d_txt
```

```python
# Read CSV
import csv
def csvtodict (csvFile, delim):
    headers = None
    content = {}
    reader=csv.reader(open(csvFile), delimiter=delim)
    i = 1
    for row in reader:
        if reader.line_num == 1:
            headers = row[1:]
        else:
            content[i] = dict(zip(headers, row[1:]))
            i = i + 1
    return content

d_csv = csvtodict(incsv, ',')
print len(d_csv)
```

```python
# read Excel
import xlrd
def xls2dictionary (infile):
    d={}
    wb = xlrd.open_workbook(infile)
    sh = wb.sheet_by_index(0) # 0 means reading the first sheet
    print sh.cell(0,0).value
    for r in range(sh.nrows):
        if r !=0:
            d[r] = {}
            for c in range(sh.ncols):
                cell_value_id = sh.cell(0,c).value # reading the header.
                d[r][cell_value_id] = sh.cell(r,c).value
    return d

d_xlsx=xls2dictionary(inexcel)
print len(d_xlsx)
```

```python
# Read file name
import os
def read_filename (indirectory, file_extension):
    listtext= []
    for f in os.listdir(indirectory):
        if f.endswith(file_extension):
            listtext.append(f)
    return listtext
listtext = read_filename(indirectory,'.adt')
print len(listtext)
```

```python
# Read table in database
import pyodbc
def read_database_table (dsn,schema,select, frm, whr):
    con_string = 'dsn=%s; schema=%s; dbmax_text=32767;trusted_connection=yes' % (dsn, schema)
    conn = pyodbc.connect(con_string)
    cursor = conn.cursor()
    cursor.execute("select %s from %s where %s " % (select, frm, whr))
    rows = cursor.fetchall()
    return rows
d_db = read_database_table("SurveyTrak11",'dba',"blInUse, vProjectId, vTable, vColumn, lvColumnDescription",
                          "DBA.tDataDictionary","vProjectId = 'SRC.SRO.PSID15.TEST3'")
print len(d_db)
```

```python
# export from dictionary to text
import csv
def export_Dict2CSV (d_FPR, header, CSVoutfile):
    f = open(CSVoutfile, 'wb')
    cw = csv.writer(f, delimiter=',', quotechar='"')
    # Write headers #
    cw.writerow(header)
    # Write data to output file #
    for key in d_FPR.keys() :
        outdata = []
        for subkey in header:
            outdata.append(d_FPR[key][subkey])
        cw.writerow(outdata)
export_Dict2CSV(d_csv, d_csv[5].keys(), outtext)
```
